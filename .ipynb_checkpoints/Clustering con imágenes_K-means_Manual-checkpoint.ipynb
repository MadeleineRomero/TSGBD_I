# Clustering con imágenes - K-means paso a paso

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
```

### Definición de funciones

Inicializa los centroides seleccionando k puntos aleatorios del conjunto de datos:

```python
def initialize_centroids(data, k):    
    indices = np.random.choice(len(data), k, replace=False)
    centroids = data[indices]
    return centroids
```

Asigna cada punto de datos al clúster cuyo centroide es el más cercano:

```python
def assign_to_clusters(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
    clusters = np.argmin(distances, axis=1)
    return clusters
```

Actualiza los centroides calculando la media de los puntos asignados a cada clúster

```python
def update_centroids(data, clusters, k):
    centroids = np.array([np.mean(data[clusters == i], axis=0) for i in range(k)])
    return centroids
```

Verifica la convergencia comparando los nuevos centroides con los anteriores

```python
def k_means(data, k, max_iterations=100):
    centroids = initialize_centroids(data, k)

    for _ in range(max_iterations):
        clusters = assign_to_clusters(data, centroids)
        new_centroids = update_centroids(data, clusters, k)

        
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return clusters, centroids
```

### Cargar datos

```python
digits = load_digits()
data = digits.data
```

```python
print(data.shape)
```

Preprocesamiento (invertir los valores para una mejor visualización en escala de grises)

```python
data = 255-data
np.random.seed(1)
```

### Aplicar k-means paso a paso

```python
n = 10 #Un cluster por cada número del 0 al 9
Z, centroids = k_means(data, n)
```

```python
for i in range(0,n):
    fila =np.where(Z==i)[0]
    num = fila.shape[0]
    r=num//10
    print("cluster"+str(i))
    print(str(num)+" elementos")
    
    plt.figure(figsize=(10,10))
    for k in range(0,num):
        plt.subplot(r+1,10,k+1)
        imagen = data[fila[k], ]
        imagen = imagen.reshape(8,8)
        plt.imshow(imagen,cmap=plt.cm.gray)
        plt.axis('off')
    plt.show()
```
