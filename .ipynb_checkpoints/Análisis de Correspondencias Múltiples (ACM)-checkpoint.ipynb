### * Explica qué es el ACM y en qué circunstancias es apropiado utilizar esta técnica
El Análisis de Correspondencias Múltiples (ACM) es una técnica estadística utilizada para analizar y visualizar las relaciones entre dos o más variables categóricas. Es una extensión del Análisis de Correspondencias Simple (AC), que se aplica a tablas de contingencia de dos dimensiones. El ACM es apropiado cuando se desea explorar las interrelaciones entre tres o más variables categóricas y se busca identificar patrones subyacentes o estructuras en los datos.


### * Describe cómo se relaciona el ACM con el PCA y en qué se diferencian.
El ACM está relacionado con el Análisis de Componentes Principales (PCA) en el sentido de que ambos son técnicas de reducción de dimensionalidad que transforman variables correlacionadas en un conjunto de valores no correlacionados llamados componentes principales. Sin embargo, mientras que el PCA se utiliza principalmente para datos cuantitativos, el ACM está diseñado para datos categóricos. Además, el PCA se centra en maximizar la varianza, mientras que el ACM busca maximizar la inercia, que es una medida de la calidad de la representación de los datos en un espacio de menor dimensión.

### * Explica para qué se utiliza una tabla disyuntiva completa en el contexto de ACM.
En el contexto de ACM, una tabla disyuntiva completa es una matriz binaria (1 o 0) que representa la presencia o ausencia de cada categoría en cada observación. Esta tabla es esencial para el análisis porque permite transformar datos categóricos en un formato que puede ser procesado mediante técnicas de álgebra lineal.

### * Enumera y describe las limitaciones y desafíos al realizar un ACM.
Las limitaciones y desafíos al realizar un ACM incluyen:
* Interpretación de resultados: Los resultados pueden ser difíciles de interpretar, especialmente cuando se trata de un gran número de categorías o niveles.
* Sensibilidad a las frecuencias marginales: El ACM puede ser sensible a las frecuencias marginales, lo que significa que las categorías con frecuencias muy altas o muy bajas pueden influir desproporcionadamente en los resultados.
* Elección de dimensiones: La elección del número de dimensiones a retener puede ser subjetiva y afectar la interpretación de los resultados.
* Pérdida de información: Al reducir la dimensionalidad, se puede perder información relevante.

### * Identifica y explica las fórmulas matemáticas clave que se requieren en el ACM.
* La matriz de datos X, que contiene las frecuencias observadas de cada combinación de categorías de las variables. Cada fila representa una categoría de la primera variable, y cada columna representa una categoría de la segunda variable. Por ejemplo, si tenemos dos variables: género (hombre, mujer) y color favorito (rojo, azul, verde), la matriz de datos tendrá dos filas y tres columnas, y cada elemento será el número de personas que tienen esa combinación de género y color favorito.
* La matriz de perfiles P, que se obtiene dividiendo cada elemento de la matriz de datos por el total de la fila correspondiente. Esta matriz representa las proporciones de cada categoría de la segunda variable dentro de cada categoría de la primera variable. Por ejemplo, si en la matriz de datos tenemos que hay 10 hombres que prefieren el rojo, 20 que prefieren el azul y 15 que prefieren el verde, la primera fila de la matriz de perfiles será (0.2,0.4,0.3), que indica que el 20% de los hombres prefieren el rojo, el 40% el azul y el 30% el verde.
* La matriz de pesos D<sub>r</sub>, que se obtiene dividiendo cada elemento de la matriz de datos por el total de la columna correspondiente. Esta matriz representa las proporciones de cada categoría de la primera variable dentro de cada categoría de la segunda variable. Por ejemplo, si en la matriz de datos tenemos que hay 10 hombres y 15 mujeres que prefieren el rojo, 20 hombres y 25 mujeres que prefieren el azul y 15 hombres y 20 mujeres que prefieren el verde, la primera columna de la matriz de pesos será (0.4,0.6), que indica que el 40% de los que prefieren el rojo son hombres y el 60% son mujeres.
* La matriz de inercias M, que se obtiene multiplicando la matriz de perfiles por la matriz de pesos. Esta matriz representa la contribución de cada combinación de categorías a la variabilidad total de los datos. Por ejemplo, si en la matriz de perfiles tenemos que la primera fila es (0.2,0.4,0.3) y en la matriz de pesos tenemos que la primera columna es (0.4,0.6), el primer elemento de la matriz de inercias será 0.2×0.4=0.08, que indica que el 8% de la inercia total se debe a la combinación de hombres y rojo.
* La inercia total I, que se obtiene sumando todos los elementos de la matriz de inercias. Esta medida representa la variabilidad total de los datos, y se puede interpretar como una generalización de la varianza para variables cualitativas. Por ejemplo, si la matriz de inercias tiene los elementos (0.08,0.16,0.12), (0.12,0.25,0.18) y (0.1,0.2,0.15), la inercia total será 0.08+0.16+0.12+0.12+0.25+0.18+0.1+0.2+0.15=1.36.
* La descomposición en valores singulares (SVD) de la matriz de perfiles, que se obtiene expresando la matriz de perfiles como el producto de tres matrices: P=UΣV<sup>T</sup>, donde U y V son matrices ortogonales que contienen los vectores propios de PP<sup>T</sup> y P<sup>T</sup>P, respectivamente, y Σ es una matriz diagonal que contiene los valores singulares de P, que son la raíz cuadrada de los valores propios de PP<sup>T</sup> o P<sup>T</sup>P. Esta descomposición permite reducir la dimensionalidad de los datos, proyectando las filas y las columnas de la matriz de perfiles en un espacio de menor dimensión, donde los ejes principales son los vectores propios de mayor valor singular.
* Los puntajes factoriales de las filas y las columnas, que se obtienen multiplicando la matriz de perfiles por los vectores propios de P<sup>T</sup>P o por los vectores propios de PP<sup>T</sup>, respectivamente. Estos puntajes representan las coordenadas de cada categoría de las variables en el espacio factorial, y permiten visualizar las relaciones entre las categorías mediante un gráfico de biplot.
* La inercia explicada por cada eje factorial, que se obtiene dividiendo el valor singular al cuadrado correspondiente por la inercia total. Esta medida representa la proporción de variabilidad de los datos que se explica por cada eje factorial, y se puede interpretar como una generalización de la varianza explicada por cada componente principal.



### * Detalla cómo deben interpretarse los gráficos resultantes de un ACM, incluyendo un gráfico de ejemplo y su explicación.
Al interpretar los gráficos resultantes de un ACM, como el biplot con representación habitual, se presta atención a la dirección, magnitud y ángulo de los vectores para comprender las relaciones entre variables, la variabilidad relativa y la contribución de cada variable a la posición de las observaciones en el espacio bidimensional del biplot.

En la representación habitual de un biplot, donde las escalas no se colocan directamente en los ejes, sino que las variables se representan como vectores, la interpretación se centra en la dirección y la magnitud de estos vectores:

* Dirección del vector: La dirección del vector indica la relación entre la variable y la orientación de los ejes oblicuos del biplot. Un vector que apunta en la misma dirección que otro sugiere una correlación positiva entre esas dos variables, mientras que una dirección opuesta sugiere una correlación negativa.

* Magnitud del vector: La longitud del vector es proporcional a la variabilidad de la variable representada. Vectores más largos indican mayor variabilidad en esa dirección. Si dos vectores son de longitudes similares, las variables que representan tienen una variabilidad comparable.

* Ángulo entre vectores: El ángulo entre dos vectores refleja la relación entre las variables que representan. Un ángulo pequeño indica una correlación positiva, mientras que un ángulo grande sugiere una correlación negativa.

* Proyección de puntos en vectores: La proyección perpendicular de un punto de datos sobre un vector proporciona información sobre el valor de esa variable para esa observación. La posición relativa de los puntos en relación con los vectores ayuda a interpretar la contribución de cada variable a la posición de una observación en el espacio definido por el biplot.

### * Discute el papel que juega la varianza en los métodos de ACM y PCA.
La varianza juega un papel crucial tanto en ACM como en PCA, ya que ambos métodos buscan resumir la variabilidad en los datos. En el PCA, la varianza se maximiza para encontrar las direcciones de mayor variabilidad, mientras que en el ACM, la inercia (una medida relacionada con la varianza) se maximiza para encontrar las dimensiones que mejor representan las asociaciones entre categorías.

### * Describe la representación matricial de la varianza en ACM y PCA.
La representación matricial de la varianza en ACM y PCA implica la descomposición de la matriz de datos en componentes que explican la mayor parte de la variabilidad (varianza en PCA, inercia en ACM). En PCA, esto se hace a través de la descomposición en valores propios de la matriz de covarianza, mientras que en ACM se utiliza la SVD de la matriz de perfiles.

### Frecuencia de las variables

```python
# Eliminar las columnas 'Raza' y 'Funcion'
ds = data.drop(['Raza', 'Funcion'], axis=1)

# Inicializar un diccionario para guardar las frecuencias por columna
frecuencias = {column: ds[column].value_counts() for column in ds}

# Convertir el diccionario en un DataFrame, llenando valores faltantes con 0 y convirtiendo a enteros
tabla_frecuencias = pd.DataFrame(frecuencias).fillna(0).astype(int)

# Transponer la tabla para que las columnas sean atributos
tabla_frecuencias = tabla_frecuencias.transpose()

# Resetear el índice y renombrar las columnas para que 'Característica' sea una columna
tabla_frecuencias.reset_index(inplace=True)
tabla_frecuencias.rename(columns={'index': 'Característica'}, inplace=True)

# Imprimir la tabla de frecuencias resultante
print(tabla_frecuencias)
```

### Diagramas de barras

```python
# Define el número de gráficos por columna y fila
nrows = 2
ncols = 3

# Crea la figura y los ejes para los subplots
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(10, 8))

# Lista de características a graficar (a elegir de tu DataFrame 'ds')
caracteristicas = ds.columns.tolist()[:6]

# Recorre las características y crea un gráfico de barras para cada una
for i, caracteristica in enumerate(caracteristicas):
    row = i // ncols
    col = i % ncols
    ax = axes[row, col]
    frecuencias_ordenadas = ds[caracteristica].value_counts().sort_index()
    ax.bar(frecuencias_ordenadas.index, frecuencias_ordenadas.values)
    ax.set_title(f'{caracteristica}')
    ax.set_ylabel('Conteo de Frecuencias')

for i in range(len(caracteristicas), nrows*ncols):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```

Genere una nube de puntos, ¿hay similitudes entre razas de perros? Y si las hay, ¿Qué razas presentan similitudes según sus características? ¿Qué porcentaje de la varianza se explica en el primer plano factorial?

```python
qualitative_columns = ['Raza','Tamaño', 'Peso', 'Velocidad', 'Inteligencia', 'Afectividad', 'Agresividad']
dog_breeds_qualitative = data[qualitative_columns]
```

### Realiza el análisis de correspondencia múltiple (MCA)

```python
mca = prince.MCA(n_components=10, n_iter=3, copy=True, check_input=True, engine='sklearn', random_state=42)
mca = mca.fit(data.drop(columns=['Raza','Funcion']))
```

```python
sns.set_theme(style="white")

individual_coordinates = mca.row_coordinates(dog_breeds_qualitative)

# Crea un gráfico de dispersión usando las coordenadas de los individuos.
plt.figure(figsize=(12, 10))
plt.scatter(individual_coordinates[0], individual_coordinates[1], alpha=0.8)

# Añade etiquetas a cada punto que representará cada raza.
if 'Raza' in dog_breeds_qualitative:
    for i, raza in enumerate(dog_breeds_qualitative['Raza']):
        plt.text(individual_coordinates.iloc[i, 0], individual_coordinates.iloc[i, 1], raza)

# Personaliza los ejes, títulos y otros elementos del gráfico.
plt.xlabel('Dim 1')
plt.ylabel('Dim 2')
plt.title('Representación Individual de las Razas')
plt.axvline(0, color='grey', lw=1)
plt.axhline(0, color='grey', lw=1)
plt.grid(True)
plt.show()
```

Podemos identificar con precisión razas caninas como el Pekinés y el Chihuahua, que a menudo presentan notables similitudes debido a su estatura reducida y ciertas características físicas compartidas. Del mismo modo, entre las razas de mayor tamaño, como el Boxer, el Dálmata y el Labrador, se pueden observar atributos físicos comunes que facilitan su reconocimiento y comparación. Estas semejanzas abarcan desde estructura corporal hasta rasgos temperamentales propios de su estirpe.

```python
eigenvalues = mca.eigenvalues_

# Calculate the percentage of variance for each eigenvalue
percentage_of_variance = (eigenvalues / np.sum(eigenvalues)) * 100

# Calculate cumulative variance
cumulative_variance = np.cumsum(percentage_of_variance)

# Create a DataFrame with eigenvalues and their percentage of variance
eigenvalues_df = pd.DataFrame({
    'Eigenvalue': eigenvalues,
    '% Variance': percentage_of_variance,
    '% Cumulative Variance': cumulative_variance
})

print(eigenvalues_df)
```

```python
# Create a screeplot
plt.figure(figsize=(12, 10))
plt.bar(range(1, len(eigenvalues) + 1), percentage_of_variance, color='skyblue', alpha=0.7, align='center')
plt.title('Screeplot - Percentage of Variance')
plt.xlabel('Dimensions')
plt.ylabel('% Variance')
plt.xticks(range(1, len(eigenvalues) + 1))

# Add a red dashed line at a specific threshold
plt.axhline(y=7.14, linestyle='--', color='red', label='Threshold')

# Add labels for each bar
for i, value in enumerate(percentage_of_variance):
    plt.text(i + 1, value + 0.2, f'{value:.2f}%', ha='center', va='bottom')

# Show the plot with adjusted y-axis limit
plt.ylim(0, max(15, np.max(percentage_of_variance) + 1))  # Set the y-axis limit
plt.legend()
plt.show()
```

El análisis de las componentes principales nos proporciona información valiosa sobre la distribución de la varianza en nuestros datos. Al examinar los porcentajes de varianza explicada por cada componente, podemos determinar cuánta información se captura al incluir un número específico de componentes principales.

En este caso, observamos que los primeros 6 componentes principales explican conjuntamente el 90.49% de la varianza total. Este resultado sugiere que al retener estos 6 componentes principales, logramos una representación significativa de la variabilidad presente en nuestros datos.

El scree plot, que visualiza la contribución relativa de cada componente principal, respalda esta decisión al mostrar que los porcentajes de varianza para los primeros 6 componentes son sustancialmente mayores que los de los componentes restantes. Este enfoque nos permite seleccionar un número óptimo de componentes principales que retienen la mayor cantidad posible de información relevante, simplificando así la complejidad de nuestros datos.
